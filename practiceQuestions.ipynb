# 1. FIND REVERSE OF THE LIST #

```python


list=[8,77,99,33,22,44]

rev_list=[]                                                                                         #list[::-1]

for i in list:
    
    rev_list.insert(0,i)
    
print(rev_list)
```

# 2. SUM OF LIST #

```python

list=[8,77,99,33,22,44]

add=list[0]+list[1]+list[2]+list[3]+list[4]+list[5]

print(f"sum of list  : {list}  is  {add}")
```

# 2. AVERAGE OF LIST #

```python
avg=add/6
print(f"average of list {list} is :  {avg}")
```

# FIND UNIQUE VALUES + COUNT ORDERS #

```python

customer_orders=["mango","dates","apple",
                 "pineapple","apple","dates",
                 "mango","dates","apple"]
unique_orders=list(set(customer_orders))
print(unique_orders)
```

```python

customer_orders = ["mango","dates","apple",
                   "pineapple","apple","dates",
                   "mango","dates","apple"]

order_count = {}

for i in customer_orders:
    if i in order_count:
        order_count[i] +=1
    else:
        order_count[i] = 1

print(order_count)
```

# 4. REMOVE DUPLICATES #

```python

student_id=[11,22,33,11,22,33,44,66,66]

unique_val=list(set(student_id))   #CHANGING LIST TO SET ,WHICH HANDLES UNIQUE OUTPUT

print("the list without duplicates is :",unique_val)
```

# 5. EVEN VALUES FROM THE LIST #

```python
student_id=[11,22,33,11,22,33,44,66,66]

for i in student_id:
    if(i%2==0):
        print("even number is :",i)
```

```python
student_id=[11,22,33,11,22,33,44,66,66]
unique=list(set(student_id))
for i in unique:
    if (i%2==0): 
        print(f"even and unique number in list is : {i}")
```
